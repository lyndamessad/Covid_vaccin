### Fonctions permettant d'obtenir une data frame adapté au réseau : 

```python
# Cette fonction permet de convertir une list en array (lisible par le réseau) :
def pandas_list_to_array(df):
    """
    Input: dataframe of shape (x, y), containing list of length l
    Return: np.array of shape (x, l, y)
    """
    
    return np.transpose(
        np.array(df.values.tolist()),
        (0, 2, 1)
    )
```

```python
# Fonction pertmettant la conversion en array et le mapping des valeurs en integer (encodage): 

def preprocess_inputs(df, token2int, cols=['sequence', 'structure', 'predicted_loop_type']):
    """
    Input : df : trainning or test dataset
            token2int : dictionary of mapping
    Return : a mapped array 
    """
    return pandas_list_to_array(
        df[cols].applymap(lambda seq: [token2int[x] for x in seq])
    )
```

### Déclaration des colonnes à prédire : 

```python
pred_cols = ['reactivity', 'deg_Mg_pH10', 'deg_Mg_50C', 'deg_pH10', 'deg_50C']
```

#### Seules 3 colonnes vont finalement être prédit (reactivity, degMgpH10, and degMg50C)

#### Les 3 colonnes servant à l'apprentissage (sequence,structure,predicted_loop_type) doivent être convertis en integer pour être apprisent par le réseau : 

```python
# Création du dictionnaire de mapping : 
token2int = {x:i for i, x in enumerate('().ACGUBEHIMSX')}
print(token2int)

# Conversion du jeu de train en array mappée grâce aux fonctions précédentes : 
train_inputs = preprocess_inputs(train, token2int)

# conversion des lables en array :
train_labels = pandas_list_to_array(train[pred_cols])
```

### Initilisation des random seeds :

```python
tf.random.set_seed(2020)
np.random.seed(2020)
# Genérateur de valeurs aléatoires : 
y_true = tf.random.normal((32, 68, 3))
y_pred = tf.random.normal((32, 68, 3))
```

### Implémentation de la MCRMSE (metric) :

```python
def MCRMSE(y_true, y_pred):
    colwise_mse = tf.reduce_mean(tf.square(y_true - y_pred), axis=1)
    return tf.reduce_mean(tf.sqrt(colwise_mse), axis=1)
```

### Définition des jeux de train et de validation : 

```python
x_train, x_val, y_train, y_val = train_test_split(
    train_inputs, train_labels, test_size=.1, random_state=34, stratify=train.SN_filter)
```

#### La MCRMSE est une moyenne de toutes les valeurs RMSE pour chacune de nos colonnes. Cela permet d'utiliser une mesure d'évaluation à un seul chiffre, même dans le cas de sorties multiples.

```python
public_df = test.query("seq_length == 107")
private_df = test.query("seq_length == 130")

public_inputs = preprocess_inputs(public_df, token2int)
private_inputs = preprocess_inputs(private_df, token2int)
```

### Réseau récurrent GRU : 

```python
def gru_layer(hidden_dim, dropout):
    return L.Bidirectional(L.GRU(
        hidden_dim, dropout=dropout, return_sequences=True, kernel_initializer='orthogonal'))
```

```python
def build_model(embed_size, seq_len=107, pred_len=68, dropout=0.5, 
                sp_dropout=0.2, embed_dim=100, hidden_dim=150, n_layers=3):
    inputs = L.Input(shape=(seq_len, 3))
    embed = L.Embedding(input_dim=embed_size, output_dim=embed_dim)(inputs)
    
    reshaped = tf.reshape(
        embed, shape=(-1, embed.shape[1],  embed.shape[2] * embed.shape[3])
    )
    hidden = L.SpatialDropout1D(sp_dropout)(reshaped)
    
    for x in range(n_layers):
        hidden = gru_layer(hidden_dim, dropout)(hidden)
    
    # Troncage des séquence en 68 bases :
    truncated = hidden[:, :pred_len]
    out = L.Dense(5, activation='linear')(truncated)
    
    model = tf.keras.Model(inputs=inputs, outputs=out)
    model.compile(tf.optimizers.Adam(), loss=MCRMSE,metrics=['accuracy'])
    
    return model
```

```python
from tensorflow.keras.utils import plot_model
```

```python
model = build_model(embed_size=len(token2int))
model.summary()
```

```python
#plot_model(model, show_shapes=True)
```

```python
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    batch_size=32,
    epochs=80,
    verbose=2,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(patience=5),
        tf.keras.callbacks.ModelCheckpoint('gru.h5')
    ]
)
```

```python
fig = px.line(
    history.history, y=['loss', 'val_loss'],
    labels={'index': 'epoch', 'value': 'MCRMSE'}, 
    title='Training History')
fig.show()
```

```python
fig = px.line(
    history.history, y=['accuracy', 'val_accuracy'],
    labels={'index': 'epoch', 'value': 'accuracy'}, 
    title='Training History')
fig.show()
```

```python
model_public = build_model(seq_len=107, pred_len=107, embed_size=len(token2int))
model_private = build_model(seq_len=130, pred_len=130, embed_size=len(token2int))

model_public.load_weights('gru.h5')
model_private.load_weights('gru.h5')
```

```python
public_preds = model_public.predict(public_inputs)
private_preds = model_private.predict(private_inputs)
```

```python
preds_ls = []

for df, preds in [(public_df, public_preds), (private_df, private_preds)]:
    for i, uid in enumerate(df.id):
        single_pred = preds[i]

        single_df = pd.DataFrame(single_pred, columns=pred_cols)
        single_df['id_seqpos'] = [f'{uid}_{x}' for x in range(single_df.shape[0])]

        preds_ls.append(single_df)

preds_df = pd.concat(preds_ls)
preds_df.head()
```

```python
submission = sample_df[['id_seqpos']].merge(preds_df, on=['id_seqpos'])
submission.to_csv('submission.csv', index=False)
```
